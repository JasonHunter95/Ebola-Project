This python notebook uses shape data from [Natural Earth Data](https://www.naturalearthdata.com/) to aid in the visualization of pandemic data.

```python
import geopandas as gpd
# Load the data
geodata = gpd.read_file("data/shapes/ne_10m_admin_0_countries.shp")
print(geodata.head())
print(geodata.columns)
```

```python
Continents = geodata['CONTINENT'].unique()
print(Continents)

Countries = geodata['ADMIN'].unique()
print(Countries)

Cities = geodata['NAME'].unique()
print(Cities)

# Filter the data for a specific continent and save to a new shapefile
# each shapefile will contain just one continent and the countries in that continent
Africa = geodata[geodata['CONTINENT'] == 'Africa']
print(Africa)
print(Africa['ADMIN'].unique())
Africa.to_file("data/shapes/africanCountries/Africa.shp")

Antarctica = geodata[geodata['CONTINENT'] == 'Antarctica']
print(Antarctica)
print(Antarctica['ADMIN'].unique())
Antarctica.to_file("data/shapes/antarcticanCountries/Antarctica.shp")

Asia = geodata[geodata['CONTINENT'] == 'Asia']
print(Asia)
print(Asia['ADMIN'].unique())
Asia.to_file("data/shapes/asianCountries/Asia.shp")

Oceania = geodata[geodata['CONTINENT'] == 'Oceania']
print(Oceania)
print(Oceania['ADMIN'].unique())
Oceania.to_file("data/shapes/oceanianCountries/Oceania.shp")

Europe = geodata[geodata['CONTINENT'] == 'Europe']
print(Europe)
print(Europe['ADMIN'].unique())
Europe.to_file("data/shapes/europeanCountries/Europe.shp")

NorthAmerica = geodata[geodata['CONTINENT'] == 'North America']
print(NorthAmerica)
print(NorthAmerica['ADMIN'].unique())
NorthAmerica.to_file("data/shapes/northAmericanCountries/NorthAmerica.shp")

SouthAmerica = geodata[geodata['CONTINENT'] == 'South America']
print(SouthAmerica)
print(SouthAmerica['ADMIN'].unique())
SouthAmerica.to_file("data/shapes/southAmericanCountries/SouthAmerica.shp")

## add seven seas

```

```python
# Now lets create a new shapefile for each country in the continent
# Africa
for country in Africa['ADMIN'].unique():
    country_data = Africa[Africa['ADMIN'] == country]
    print(country_data)
    print(country_data['ADMIN'].unique())
    country_data.to_file(f"data/shapes/africanCountries/{country}.shp")
```

```python
# Lets map the countries in Africa
import matplotlib.pyplot as plt
import geopandas as gpd
# Load the data
africa = gpd.read_file("data/shapes/africanCountries/Africa.shp")
# Plot the data
africa.plot()
plt.title("African Countries")
plt.show()
```
